## 1. Why Hypothesis Testing Matters

Hypothesis testing helps us answer questions like:
- Is this result real or due to randomness?
- Is model A better than model B?
- Did a change actually improve performance?

In research, hypothesis testing supports **evidence-based decisions**.


```python

import numpy as np
import matplotlib.pyplot as plt
```


## 2. Null and Alternative Hypotheses

- **Null hypothesis (H‚ÇÄ)**: default assumption (no effect, no difference)
- **Alternative hypothesis (H‚ÇÅ)**: what we want to test (effect exists)

Example:
- H‚ÇÄ: New model performs the same as old model
- H‚ÇÅ: New model performs better



## 3. Test Statistic (Intuition)

A **test statistic** summarizes data into a single number.
It tells us how far our observation is from what H‚ÇÄ predicts.

We do NOT focus on formulas here.



## 4. p-value (Very Important)

The **p-value** answers:
> If H‚ÇÄ were true, how surprising is the observed result?

- Small p-value ‚Üí unlikely under H‚ÇÄ
- Large p-value ‚Üí consistent with H‚ÇÄ

Typical threshold:
- p < 0.05 ‚Üí reject H‚ÇÄ


```python

# Simulating p-value intuition
np.random.seed(0)

# Assume H0: mean = 0
population = np.random.normal(0, 1, 100000)

# Sample with a shift
sample = np.random.normal(0.3, 1, 50)

observed_mean = sample.mean()

# Sampling distribution under H0
means = [np.mean(np.random.choice(population, 50)) for _ in range(5000)]

plt.hist(means, bins=40, density=True)
plt.axvline(observed_mean, color='red', linestyle='--')
plt.title("Sampling Distribution under H‚ÇÄ")
plt.show()
```


## 5. Type I and Type II Errors

- **Type I error**: reject H‚ÇÄ when it is true (false positive)
- **Type II error**: fail to reject H‚ÇÄ when it is false (false negative)

Tradeoff exists between the two.



## 6. Significance Level (Œ±)

- Œ± = probability of Type I error
- Common choice: Œ± = 0.05

Lower Œ± ‚Üí more conservative tests



## 7. A/B Testing (ML & Industry)

A/B testing compares two versions:
- Model A vs Model B
- Feature A vs Feature B

Hypothesis testing tells us:
- Whether observed improvement is statistically meaningful


```python

# Simple A/B testing simulation
A = np.random.normal(0.5, 0.1, 100)
B = np.random.normal(0.55, 0.1, 100)

print("Mean A:", A.mean())
print("Mean B:", B.mean())
```
