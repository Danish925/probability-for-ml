## 1. Why Statistics Matters in ML

Statistics helps us:
- Summarize data
- Understand variability
- Detect relationships
- Identify noise and outliers

In research, **statistics guides correct conclusions** from data.


```python

import numpy as np
import matplotlib.pyplot as plt
```


## 2. Measures of Central Tendency

### Mean
Average value of data.

### Median
Middle value (robust to outliers).

### Mode
Most frequent value.


```python

data = np.array([2, 3, 4, 5, 100])

print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Mode (approx):", np.bincount(data).argmax() if data.max() < 1000 else "Not applicable")
```


## 3. Measures of Spread

### Variance
Average squared deviation from mean.

### Standard Deviation
Square root of variance (same unit as data).


```python

data = np.random.normal(0, 2, 1000)
print("Variance:", data.var())
print("Standard Deviation:", data.std())
```


## 4. Covariance

Covariance measures how **two variables change together**.

- Positive: increase together
- Negative: move in opposite directions
- Zero: no linear relationship


```python

x = np.random.normal(0, 1, 1000)
y = 2*x + np.random.normal(0, 0.5, 1000)

cov = np.cov(x, y)
print("Covariance matrix:\n", cov)
```


## 5. Correlation

Correlation is **normalized covariance**.

Range:
- âˆ’1 â†’ perfect negative
- 0 â†’ no linear relation
- +1 â†’ perfect positive


```python

corr = np.corrcoef(x, y)
print("Correlation matrix:\n", corr)
```


## 6. Covariance vs Correlation (Important)

- Covariance depends on scale
- Correlation is scale-invariant

ML prefers **correlation for feature relationships**.



## 7. Outliers

Outliers are extreme values that:
- Skew mean
- Affect variance
- Can mislead models


```python

data = np.concatenate([np.random.normal(0,1,1000), np.array([10, 12, -11])])
plt.hist(data, bins=50)
plt.title("Outliers in Data")
plt.show()
```


## 8. Data Visualization (Statistical Insight)

Visuals help detect:
- Skewness
- Outliers
- Distribution shape


```python

plt.boxplot(data)
plt.title("Boxplot Showing Outliers")
plt.show()
```


## 9. ML Connection Summary

- Mean/variance â†’ feature scaling
- Covariance/correlation â†’ feature selection
- Outliers â†’ robust modeling
- Visualization â†’ sanity checks before modeling
