## 1. What is a Random Variable?

A **random variable** assigns a numerical value to each outcome of a random experiment.

Examples:
- Coin toss â†’ 0 or 1
- Dice roll â†’ 1 to 6
- Height of a person

### ML intuition
Data in ML is modeled as outcomes of random variables.


```python

import numpy as np
import matplotlib.pyplot as plt
```


## 2. Discrete vs Continuous Random Variables

- **Discrete**: countable values (coin toss, dice)
- **Continuous**: infinite range (height, weight)


```python

# Discrete example: dice simulation
dice = np.random.randint(1, 7, size=10000)
values, counts = np.unique(dice, return_counts=True)

for v, c in zip(values, counts):
    print(v, ":", c/len(dice))
```


## 3. Probability Mass Function (PMF)

PMF gives probability for discrete random variables:
P(X = x)


```python

pmf = counts / counts.sum()
plt.bar(values, pmf)
plt.xlabel("Value")
plt.ylabel("Probability")
plt.title("PMF of Dice Roll")
plt.show()
```


## 4. Probability Density Function (PDF)

PDF is used for continuous random variables.
Probability is area under the curve.


```python

data = np.random.normal(0, 1, 10000)
plt.hist(data, bins=50, density=True)
plt.title("Histogram approximating PDF")
plt.show()
```


## 5. Cumulative Distribution Function (CDF)

CDF is:
F(x) = P(X â‰¤ x)


```python

sorted_data = np.sort(data)
cdf = np.arange(1, len(sorted_data)+1) / len(sorted_data)
plt.plot(sorted_data, cdf)
plt.title("Empirical CDF")
plt.show()
```


## 6. Expectation and Variance

Expectation = average value  
Variance = spread


```python

print("Mean:", data.mean())
print("Variance:", data.var())
print("Std Dev:", data.std())
```


## 7. Law of Large Numbers

As sample size increases, sample mean approaches true mean.


```python

means = [np.mean(np.random.normal(0,1,n)) for n in range(10,5000,50)]
plt.plot(range(10,5000,50), means)
plt.axhline(0, color='red')
plt.title("Law of Large Numbers")
plt.show()
```
