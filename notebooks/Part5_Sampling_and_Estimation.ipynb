## 1. Population vs Sample

- **Population**: entire set of interest (usually unknown)
- **Sample**: subset of population (what we actually observe)

In ML, we **never** see the full population.
We always learn from samples.


```python

import numpy as np
import matplotlib.pyplot as plt
```


## 2. Sampling

Sampling means selecting data points from a population.

Good sampling:
- Is random
- Represents the population well

Bad sampling:
- Introduces bias


```python

# Simulated population
population = np.random.normal(50, 10, 100000)

# Random sample
sample = np.random.choice(population, size=1000, replace=False)

print("Population mean:", population.mean())
print("Sample mean:", sample.mean())
```


## 3. Sampling Bias

Sampling bias occurs when samples are **not representative**.

This leads to **wrong conclusions**, even with large data.


```python

# Biased sampling (only large values)
biased_sample = population[population > 60][:1000]

print("Biased sample mean:", biased_sample.mean())
```


## 4. Estimation

An **estimator** is a rule that estimates a population quantity.

Examples:
- Sample mean ‚Üí estimates population mean
- Sample variance ‚Üí estimates population variance


```python

# Estimating population mean using samples
estimates = []
for _ in range(200):
    sample = np.random.choice(population, size=500, replace=False)
    estimates.append(sample.mean())

plt.hist(estimates, bins=30)
plt.axvline(population.mean(), color='red', linestyle='--')
plt.title("Sampling Distribution of the Mean")
plt.show()
```


## 5. Bias of an Estimator

- **Unbiased estimator**: expected estimate equals true value
- **Biased estimator**: systematically off

Bias affects model accuracy.



## 6. Variance of an Estimator

Variance measures **how much estimates vary** between samples.

High variance:
- Unstable estimates
- Overfitting in ML



## 7. Bias‚ÄìVariance Tradeoff ‚≠ê‚≠ê‚≠ê

Total error = Bias¬≤ + Variance + Noise

- High bias ‚Üí underfitting
- High variance ‚Üí overfitting

This is one of the **most important concepts in ML research**.


```python

# Visual intuition using sample size
sample_sizes = [10, 50, 200, 1000]
errors = []

for n in sample_sizes:
    means = [np.mean(np.random.choice(population, n, replace=False)) for _ in range(200)]
    errors.append(np.var(means))

plt.plot(sample_sizes, errors, marker='o')
plt.xlabel("Sample size")
plt.ylabel("Variance of estimator")
plt.title("Estimator Variance vs Sample Size")
plt.show()
```


## 8. Confidence Intervals (Intuition Only)

A confidence interval gives a **range** where the true parameter likely lies.

We avoid heavy formulas.
Just remember:
- More data ‚Üí narrower interval
- More noise ‚Üí wider interval



## 9. ML Connection Summary

- Sampling explains train/test splits
- Estimation explains parameter learning
- Bias‚Äìvariance explains model selection
- Confidence explains uncertainty in predictions
