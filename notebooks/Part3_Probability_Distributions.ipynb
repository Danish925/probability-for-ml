## 1. Why Probability Distributions Matter in ML

A **probability distribution** describes how values of a random variable are spread.

In ML, distributions are used to:
- Model data generation
- Model noise and errors
- Make assumptions in algorithms
- Perform probabilistic inference


```python

import numpy as np
import matplotlib.pyplot as plt
```


## 2. Bernoulli Distribution

Used for **binary outcomes**.

Examples:
- Coin toss (0 or 1)
- Spam vs not spam

Parameter:
- p = probability of success (1)

Mean = p  
Variance = p(1 âˆ’ p)


```python

p = 0.7
samples = np.random.binomial(1, p, size=10000)
values, counts = np.unique(samples, return_counts=True)

plt.bar(values, counts / counts.sum())
plt.xticks([0,1])
plt.title("Bernoulli Distribution")
plt.show()
```


## 3. Binomial Distribution

Models **number of successes in N trials**.

Parameters:
- n = number of trials
- p = success probability


```python

n, p = 10, 0.5
samples = np.random.binomial(n, p, size=10000)

plt.hist(samples, bins=range(n+2), density=True)
plt.title("Binomial Distribution")
plt.xlabel("Successes")
plt.ylabel("Probability")
plt.show()
```


## 4. Uniform Distribution

All values in an interval are equally likely.


```python

samples = np.random.uniform(0, 1, size=10000)
plt.hist(samples, bins=50, density=True)
plt.title("Uniform Distribution")
plt.show()
```


## 5. Gaussian (Normal) Distribution â­â­â­

Most important distribution in ML.

Parameters:
- Mean (Î¼)
- Standard deviation (Ïƒ)


```python

samples = np.random.normal(0, 1, size=10000)
plt.hist(samples, bins=50, density=True)
plt.title("Gaussian Distribution")
plt.show()
```


## 6. Central Limit Theorem (Intuition)

The mean of many samples tends toward a Gaussian distribution.


```python

means = [np.mean(np.random.uniform(0,1,1000)) for _ in range(5000)]
plt.hist(means, bins=50, density=True)
plt.title("Central Limit Theorem Simulation")
plt.show()
```


## 7. Exponential Distribution (Optional)

Models time between events.


```python

samples = np.random.exponential(scale=1.0, size=10000)
plt.hist(samples, bins=50, density=True)
plt.title("Exponential Distribution")
plt.show()
```
